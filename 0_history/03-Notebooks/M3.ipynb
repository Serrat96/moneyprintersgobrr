```python
from fredapi import Fred
import pandas as pd
import plotly.express as px
```

```python
def get_fred_data(api_key: str, series:str):

    fred = Fred(api_key=api_key)
    series = fred.get_series(series)
    series = pd.DataFrame(series, columns=['Value'])
    return series
```

```python
m3_usa = get_fred_data('a7eca89fdf2905baea21d67b942c9ef7', 'MABMM301USM189S').dropna()
#La serie tiene ajustes estacionales
```

```python
m3_usa.to_parquet('m3_usa.parquet')
```
